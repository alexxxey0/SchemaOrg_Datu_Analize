```python
from helpers import parse_and_count_well_described_entities

schema_org_classes = [
    "Answer",
    "Book",
    "Dataset",
    "Hospital",
    "Movie",
    "MusicAlbum",
    "QAPage",
    "Recipe",
    "School",
    "ShoppingCenter"
]

filenames = {}

filenames['Answer'] = []
for i in range(118):
    filenames['Answer'].append(rf"E:\schema_org_datasets\Answer\Answer_part_{i}.gz")
    
filenames['Book'] = []
for i in range(19):
    filenames['Book'].append(rf"E:\schema_org_datasets\Book\Book_part_{i}.gz")
    
filenames['Dataset'] = []
for i in range(1, 6):
    filenames['Dataset'].append(f"../dataset_2024_p{i}.gz")
    
filenames["Hospital"] = ["../hospital_2024_p1.gz", "../hospital_2024_p2.gz"]

filenames["Movie"] = []
for i in range(1, 13):
    filenames["Movie"].append(f"../movie_2024_p{i}.gz")
    
filenames["MusicAlbum"] = []
for i in range(5):
    filenames["MusicAlbum"].append(rf"E:\schema_org_datasets\MusicAlbum\MusicAlbum_part_{i}.gz")
    
filenames["QAPage"] = []
for i in range(12):
    filenames["QAPage"].append(rf"E:\schema_org_datasets\QAPage\QAPage_part_{i}.gz")
    
filenames["Recipe"] = []
for i in range(21):
    filenames["Recipe"].append(rf"E:\schema_org_datasets\Recipe\Recipe_part_{i}.gz")
    
filenames["School"] = ["../school_2024.gz"]

filenames["ShoppingCenter"] = ["../shopping_center_2024_p1.gz", "../shopping_center_2024_p2.gz"]
```

```python
# Thresholds for “commonly used” predicates were set manually per class.
# For each class, the top predicates were examined, and the threshold corresponds
# to the point where usage drops sharply. Only predicates with usage above the threshold
# are considered “common” for counting well-described entities.


thresholds =  {
    "Answer": 0.09,
    "Book": 0.4,
    "Dataset": 0.4,
    "Hospital": 0.6,
    "Movie": 0.3,
    "MusicAlbum": 0.3,
    "QAPage": 0.9,
    "Recipe": 0.7,
    "School": 0.4,
    "ShoppingCenter": 0.5
}

processed = ["Answer", "Book", "Dataset", "Hospital"]


for cls in schema_org_classes:
    if cls not in processed:
        parse_and_count_well_described_entities(cls, filenames[cls], thresholds[cls])
        print()
```
